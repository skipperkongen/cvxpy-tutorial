```python
import cvxpy as C
import numpy as np
import matplotlib.pyplot as plt
```

## Scalar functions

```python
X = np.linspace(0,10, 100)
y = np.vectorize(lambda x: C.sqrt(x).value)(X)
plt.plot(X,y)
```

```python
X = np.linspace(0,10, 100)
y = [C.sum_squares(x).value for x in X]
plt.plot(X,y)
```

## Element-wise functions

```python
X = np.linspace(0,10, 100)
Y = C.entr(X).value
plt.plot(X,Y)
```

```python
X = np.linspace(0,10, 100)
Y = C.exp(X).value
plt.plot(X,Y)
```
